```python

import pandas as pd

# Chargement des 3 datasets - on ne garde que la colonne J-D qui reprend la moyenne annuelle
def load_and_prepare(url, region):
    df = pd.read_csv(url, header=1)
    df = df.iloc[1:]
    df = df.replace('***', pd.NA) # remplace les *** par des nan
    df['Region'] = region
    df = df[['Year', 'J-D', 'Region']] # seules 3 colonnes sont conservées
    return df

df_global = load_and_prepare("https://data.giss.nasa.gov/gistemp/tabledata_v4/GLB.Ts+dSST.csv", "Monde")
df_north  = load_and_prepare("https://data.giss.nasa.gov/gistemp/tabledata_v4/NH.Ts+dSST.csv", "Hémisphère Nord")
df_south  = load_and_prepare("https://data.giss.nasa.gov/gistemp/tabledata_v4/SH.Ts+dSST.csv", "Hémisphère Sud")

# Fusion et nettoyage.
df_final = pd.concat([df_global, df_north, df_south], ignore_index=True)
df_final['J-D'] = pd.to_numeric(df_final['J-D'], errors='coerce')
df_final['Year'] = pd.to_numeric(df_final['Year'], errors='coerce')

# Supprimer la ligne où l'année est égale à 2025 car les valeurs de J-D sont nulles
df_final = df_final[df_final['Year'] != 2025]

df_final

```

```python
# prompt: peux tu détailler chaque ligne de code en français?

import pandas as pd


# Chargement des 3 datasets - on ne garde que la colonne J-D qui reprend la moyenne annuelle
def load_and_prepare(url, region):
    """
    Charge un fichier CSV depuis une URL, le prétraite et retourne un DataFrame pandas.

    Args:
        url (str): L'URL du fichier CSV.
        region (str): La région associée aux données (e.g., "Monde", "Hémisphère Nord").

    Returns:
        pd.DataFrame: Un DataFrame contenant les années, les températures moyennes annuelles (J-D) et la région.
    """
    df = pd.read_csv(url, header=1)  # Lit le fichier CSV, en utilisant la deuxième ligne comme en-tête.
    df = df.iloc[1:]  # Supprime la première ligne du DataFrame
    df = df.replace('***', pd.NA) # Remplace les valeurs '***' (qui représentent des données manquantes) par des NaN.
    df['Region'] = region  # Ajoute une colonne 'Region' avec la région spécifiée.
    df = df[['Year', 'J-D', 'Region']] # Garde uniquement les colonnes 'Year', 'J-D' et 'Region'.
    return df

# Charge les données pour le monde, l'hémisphère nord et l'hémisphère sud en utilisant la fonction load_and_prepare.
df_global = load_and_prepare("https://data.giss.nasa.gov/gistemp/tabledata_v4/GLB.Ts+dSST.csv", "Monde")
df_north  = load_and_prepare("https://data.giss.nasa.gov/gistemp/tabledata_v4/NH.Ts+dSST.csv", "Hémisphère Nord")
df_south  = load_and_prepare("https://data.giss.nasa.gov/gistemp/tabledata_v4/SH.Ts+dSST.csv", "Hémisphère Sud")

# Fusion et nettoyage.
df_final = pd.concat([df_global, df_north, df_south], ignore_index=True) # Combine les trois DataFrames en un seul.
df_final['J-D'] = pd.to_numeric(df_final['J-D'], errors='coerce') # Convertit la colonne 'J-D' en numérique, en remplaçant les erreurs de conversion par des NaN.
df_final['Year'] = pd.to_numeric(df_final['Year'], errors='coerce') # Convertit la colonne 'Year' en numérique, en remplaçant les erreurs de conversion par des NaN.


# Supprimer la ligne où l'année est égale à 2025 car les valeurs de J-D sont nulles
df_final = df_final[df_final['Year'] != 2025]

df_final
```
